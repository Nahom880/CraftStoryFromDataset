Starbucks, a global coffeehouse chain renowned for its premium coffee and inviting atmosphere, operates over 31,000 outlets in 83 countries1. The company, established in 1971, has evolved into an iconic brand, offering a diverse menu of high-quality beverages, snacks, and pastries1.

The Starbucks menu is designed with psychological tactics to enhance the customer experience and increase sales2. ## 1. Sourcing and loading

### 1a. Import Packages

```python
import pandas as pd
import matplotlib.patches as mpatches
import dataframe_image as dfi
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import os
```

### 1b. Load data

```python
drinks_df = pd.read_csv('starbucks-menu-nutrition-drinks.csv')
food_df = pd.read_csv('starbucks-menu-nutrition-food.csv', encoding='ISO-8859-1')
menu_df = pd.read_csv('starbucks_drinkMenu_expanded.csv')
```

### 1c. Wrangling the Data 

#### 1c. 1. Checking Drinks Dataframe 

```python
drinks_df.head()
```

```python
drinks_df.info()
```

```python
drinks_df = drinks_df.rename(columns = {'Unnamed: 0':'Drink', 'Protein':'Protein (g)', 'Sodium':'Sodium (g)'})
```

```python
drinks_df = drinks_df.replace('-', np.nan)
```

```python
drinks_df = drinks_df.dropna()
```

```python
drinks_df[['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)', 'Sodium (g)']] = \
drinks_df[['Calories', 'Fat (g)', 'Carb. (g)', 'Fiber (g)', 'Protein (g)', 'Sodium (g)']].astype(float)
```

```python
drinks_df.describe()
```

#### 1c. 1. Checking Food Dataframe

```python
food_df = food_df.rename(columns = {'Unnamed: 0':'Food'})
```

```python
food_df = food_df.replace('-', np.nan)
```

### <center> Based on Beverage Category to understand the content of the drink </center> 

```python
BevCatMean = menu_df.groupby('Beverage Category').mean(numeric_only=True).astype(int)
```

```python
BevCatMean.sort_values('Calories')
```

```python
BevCatMean.columns
```

```python
BevCatMean = BevCatMean[['Calories', 'Cholesterol (mg)', \
         'Sugars (g)', 'Caffeine (mg)']]
```

```python
BevCatMean = BevCatMean.sort_values('Calories')
```

```python
BevCatMean
```

```python
BevCatMean = BevCatMean.reset_index(drop=False)
```

```python
_ = plt.figure(figsize=(10, 10))
g1 = sns.barplot(x='Calories', y="Beverage Category", orient='h',\
                 data=BevCatMean, color='b')
ax2 = plt.twiny()
g2 = sns.barplot(x='Caffeine (mg)', y='Beverage Category', orient='h',
                 data=BevCatMean, ax=ax2, color='g')

blue = mpatches.Patch(color='g', label='Caffeine')
green = mpatches.Patch(color='b', label='Calories')

plt.legend(handles=[blue, green], loc=7)

plt.show()
```
